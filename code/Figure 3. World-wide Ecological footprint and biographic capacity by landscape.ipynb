### EF and BC per capita

```python
ef_person_EF=ef_person.loc[ef_person['Record']=='EFConsPerCap'].reset_index()
ef_person_BC=ef_person.loc[ef_person['Record']=='BiocapPerCap'].reset_index()
```

```python
source_list = ef_person_EF.columns.tolist()[5:11]
source_list
```

```python
data=[]
#Aggregate Incident number by year
#new_total = crime_data.groupby('year')['IncidntNum'].sum().reset_index().rename(columns={'IncidntNum':'yeartotal'}) 
#Aggregate incident number by category by year and calculate proportion of each category per year.


for source in source_list:
    #new_df = crime_data.loc[crime_data['Category']==crime, ].groupby('year')['IncidntNum'].sum().reset_index()
    #new = new_df.merge(new_total, how = 'left', on ='year')
    #new['pert'] = new['IncidntNum']/new['yeartotal']*100
    
    trace =go.Bar(
                x=ef_person_EF['year'],
                y=ef_person_EF[source],
                name=source
        )
    data.append(trace)
    
layout = go.Layout(title = 'Figure 1.1 World Ecological Footprint per capita by source',
              xaxis = dict(title = 'Year', autotick=False),
              yaxis = dict(title = 'Ecological Footprint per capita'),
              barmode='stack',
              showlegend=True)

fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='grouped-bar-stacked')
```
