```python
data=[]

for continent in [ef_person_EF, ef_asian_total,ef_europe_total,ef_north_america_total,ef_latin_am_total,ef_africa_total, ef_oceania_total]:
    for source in source_list:
    
        trace =go.Bar(
                    x=continent['year'],
                    y=continent[source],
                    name=source
            )
        data.append(trace)
    
T_list = [True, True,True,True,True, True]
F_list = [False, False,False, False,False, False]
button_layer_1_height = 1.12
updatemenus = list([
    dict(active=-1,
         buttons=list([   
            dict(label = 'World',
                 method = 'update',
                 args = [{'visible': T_list + F_list + F_list + F_list + F_list + F_list + F_list},
                         {'title': 'Figure 1.1 World Ecological Footprint per capita by source'}]),
                        #  ,'annotations': high_annotations}]),
            dict(label = 'Asian',
                 method = 'update',
                 args = [{'visible': F_list + T_list + F_list + F_list + F_list + F_list + F_list},
                         {'title': 'Figure 1.1 Asian Ecological Footprint per capita by source'}]),
                          #'annotations': low_annotations}]),
            dict(label = 'Europe',
                 method = 'update',
                 args = [{'visible': F_list + F_list + T_list + F_list + F_list + F_list + F_list},
                         {'title': 'Figure 1.1 Europe Ecological Footprint per capita by source'}]),
                          #'annotations': low_annotations}]),
            dict(label = 'North America',
                 method = 'update',
                 args = [{'visible': F_list + F_list + F_list + T_list + F_list + F_list + F_list},
                         {'title': 'Figure 1.1 North America Ecological Footprint per capita by source'}]),   
             dict(label = 'Latin America/Caribbean',
                 method = 'update',
                 args = [{'visible': F_list + F_list + F_list + F_list + T_list + F_list + F_list},
                         {'title': 'Figure 1.1 Latin America and the Caribbean Ecological Footprint per capita by source'}]),     
             dict(label = 'Africa',
                 method = 'update',
                 args = [{'visible': F_list + F_list + F_list + F_list + F_list + T_list + F_list},
                         {'title': 'Figure 1.1 Africa Ecological Footprint per capita by source'}]),
             dict(label = 'Oceania',
                 method = 'update',
                 args = [{'visible': F_list + F_list + F_list + F_list + F_list + F_list + T_list},
                         {'title': 'Figure 1.1 Oceania Ecological Footprint per capita by source'}])               
         
         ])  
        )
    ])



layout = go.Layout(title = 'Figure 1.1 World Ecological Footprint per capita by source',
              xaxis = dict(title = 'Year', autotick=False),
              yaxis = dict(title = 'Ecological Footprint per capita'),
              barmode='stack',
              showlegend=True,
              updatemenus=updatemenus)

fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='grouped-bar-stacked')
```
